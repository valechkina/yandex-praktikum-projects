## Описание данных

Таблица `visits` (лог сервера с информацией о посещениях сайта):
- _Uid_ — уникальный идентификатор пользователя

- _Start Ts_ — дата и время начала сессии
- _End Ts_ — дата и время окончания сессии
- _Source Id_ — идентификатор рекламного источника, из которого пришел пользователь

- _Uid_ — уникальный id пользователя, который сделал заказ

- _Revenue_ — выручка Яндекс.Афиши с этого заказа

Таблица `costs` (информация о затратах на маркетинг):
- _source_id_ — идентификатор рекламного источника
- _dt_ — дата
- _costs_ — затраты на этот рекламный источник в этот день

#### 2.1.1 DAU, WAU, MAU 

Сколько людей пользуются в день, неделю, месяц? Чтобы ответить на этот вопрос, возьмем таблицу `visits` с данными о посещениях и посчитаем DAU (daily average users), WAU (weekly average users) и MAU (monthly average users).

- Чтобы посчитать недельное и месячное использование, выделим в отдельные столбцы год, месяц и неделю, а также полную дату.

```python
visits['year']  = visits['start_ts'].dt.year
visits['month'] = visits['start_ts'].astype('datetime64[M]')
visits['date'] = visits['start_ts'].dt.date
```

```python
# считаем день начала недели, за которую произошло событие. 
# далее вычитаем из даты порядковый номер дня
visits['date'] = pd.to_datetime(visits['date'])
visits['week'] = pd.to_datetime(visits['date'], unit='d') - pd.to_timedelta(visits['date'].dt.dayofweek, unit='d')
```

```python
#Проверяем, что получилось
visits.head()
```

- Посчитаем метрики. Сгруппируем данные по уникальным пользователям и найдём среднее:

```python
daily_users = visits.groupby(['date']).agg({'uid': 'nunique'}).reset_index()
weekly_users = visits.groupby(['year', 'week']).agg({'uid': 'nunique'}).reset_index()
monthly_users = visits.groupby(['year', 'month']).agg({'uid': 'nunique'}).reset_index()

daily_users.head()
```

В любом случае, здесь есть возможности для роста, особенно в весенние и летние месяцы.

- Посчитаем теперь среднее количество визитов - те же DAU, WAU и MAU - но **в среднем за весь период**.

```python
# Посчитаем метрики. Сгруппируем данные по уникальным пользователям и найдём среднее:
dau_total = visits.groupby('date').agg({'uid': 'nunique'}).mean()
wau_total = visits.groupby(['year', 'week']).agg({'uid': 'nunique'}).mean()
mau_total = visits.groupby(['year', 'month']).agg({'uid': 'nunique'}).mean()

#Смотрим какие получились средние
print('Daily Active Users (total):', int(dau_total))
print('Weekly Active Users (total):', int(wau_total)) 
print('Monthly Active Users (total):', int(mau_total))
```

- Можно также посчитать эти метрики отдельно по платформам (`device`).

```python
# определяем, какие у нас есть платформы: выделим уникальные значения из столбца `device`
devices = visits['device'].unique()

# теперь в цикле рассчитываем метрики по каждой платформе 
daily_users = []
weekly_users = []
monthly_users = []

for device in devices:
    dau = visits[visits['device'] == device].groupby('date').agg({'uid': 'nunique'}).mean()
    wau = visits[visits['device'] == device].groupby(['year', 'week']).agg({'uid': 'nunique'}).mean()
    mau = visits[visits['device'] == device].groupby(['year', 'month']).agg({'uid': 'nunique'}).mean()
    daily_users.append(int(dau))
    weekly_users.append(int(wau))
    monthly_users.append(int(mau))

# сведем получившиеся метрики по всем девайсам в одну табличку 
visits_by_device = pd.DataFrame({
    "Device": devices,
    "DAU": daily_users,
    "WAU": weekly_users,
    "MAU": monthly_users
})  

# смотрим что получилось
visits_by_device
```

Пользователей на десктопе в 2,5 раза больше, чем в приложении! Есть над чем задуматься. Возможно приложение не очень удобное, а может все планируют досуг и сидят в "Афише" в рабочее время в офисе??
Сейчас это не основной вопрос исследования, но стоит подкинуть эту информацию коллегам и продакту.

- **Как часто люди возвращаются?**

Чтобы ответить на этот вопрос, посчитаем **sticky factor**.

```python
sticky_mau = dau_total / mau_total * 100
sticky_wau = dau_total / wau_total * 100

print(sticky_wau)
print(sticky_mau)
```

"Липкость" за неделю составляет около 15.8%. Это означает, что 15 процентов пользователей заходили на сервис более 1 раза за неделю. 
А sticky factor за месяц всего 3.9% - есть над чем работать в плане удержания пользователей.

Sticky factor показывает лояльность и активность наших пользователей, проще говоря, сколько из привлеченных новых пользователей останутся и продолжат пользоваться сервисом (и потенциально совершать продажи, конечно).

**Вывод**
- Абсолютный максимум посещений за день был зафиксирован 24 ноября 2017 года.
- Довольно сильно отличаются все три метрики DAU, WAU и MAU по платформам: больше пользователей заходят на "Афишу" с десктопа.
- В среднем количество визитов было выше всего осенью 2017 и начало планомерно снижаться в 2018 году.

#### 2.1.2 Сколько раз за день пользователи в среднем заходят на сайт <a id="visits"></a>

```python
# группируем по дням: считаем общее количество пользователей (сессии) и количество уникальных пользователей
sessions_per_user = visits.groupby(['date']).agg({'uid': ['count','nunique']})

# избавляемся от мультииндексов
sessions_per_user.columns = ['n_sessions', 'n_users']

# чтобы найти количество сессий в день, разделим число сессий на количество уникальных пользователей за день
sessions_per_user['sessions_per_user_daily'] = sessions_per_user['n_sessions'] / sessions_per_user['n_users']
sessions_per_user
```

#### 2.1.3  ASL: сколько времени пользователи проводят на сайте

Рассчитаем среднюю продолжительность сессии или ASL (англ. average session length): посчитаем длину всех сессий и вычислим среднее.

```python
visits['session_duration_sec'] = (visits['end_ts'] - visits['start_ts']).dt.seconds
print('Средняя продолжительность сессии в минутах:', visits['session_duration_sec'].mean()/60)
```

Посмотрим отличается ли ASL на мобильных устройствах и на десктопе.

В среднем, на сайте пользователи проводят почти на 4 минуты дольше, чем в мобильном приложении.

Теперь построим распределение (гистограмму) и рассчитаем статистики ASL.

Наилучшим решением будет оценить типичную продолжительность сессии по моде.

```python
asl_mode = visits['session_duration_sec'].mode()
asl_mode
```

Мода равна 60 секундам. Но что конкретно это значит? Больше пользователей проводят на сайте менее минуты или чуть более? Чтобы получившееся значение было более-менее информативным, посчитаем моду не по абсолютным величинам, а по интервалам (30-45 секунд, 45-60 секунд, и т.д.)

```python
# Задаем начальную и конечную точку по минимальной и максимальной продолжительности сессии
# frequency означает длину шага, пусть будет 15 секунд
# closed right будет включать в себя начальную точку интервала и НЕ включать конечную
interval_range = pd.interval_range(start=0, freq=15, end=84480, closed='right')

# пилим столбец 'session_duration_sec' на интервалы функцией cut
visits['bin'] = pd.cut(visits['session_duration_sec'], bins=interval_range)

# смотрим, что получилось
visits.sample(5)
```

Также на статистиках мы видели, что минимальная продолжительность сессии 0, т.е. это bounce, когда пользователь зашел на страницу и сразу с нее ушел, не совершив никаких действий. Посмотрим, много ли у нас таких. Для более "традиционного" расчета возьмем все сессии короче 15 секунд, как они считаются в яндекс аналитике.

```python
len(visits[visits['session_duration_sec'] < 15]) / len(visits)
```

- Примерно 10% всех пользователей проводит на сайте менее 15 секунд.

#### 2.1.4 Retention Rate

Рассчитаем Retention Rate (коэффициент "удержания" пользователей), применяя когортный анализ.
Retention Rate показывает, сколько пользователей из когорты относительно их изначального числа вернулось в последующие периоды.

- Сначала сформируем когорты. Когортами у нас будут пользователи, зарегистрировавшиеся в течение каждого месяца за рассматриваемый период.

```python
# находим самую первую дату визита для каждого юзера, это и будет дата регистрации
first_visit = visits.groupby(['uid'])['start_ts'].min()
first_visit.name = 'first_visit'

# мержим дату регистрации в исходный датафрейм
visits = visits.join(first_visit, on='uid', how='left')

visits.head()
```

```python
# выделим из столбцов start_ts и registration_date месяцы
visits['first_visit_month'] = visits['first_visit'].astype('datetime64[M]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')

# посчитаем lifetime когорт
visits['cohort_lifetime'] = visits['session_month'] - visits['first_visit_month']

# приведем lifetime к целым значениям в месяцах
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
visits.head()
```

Ну вот, все готово для когортного анализа. Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на каждый месяц ее существования.

```python
# группируем данные по когортам (registration_month) и считаем количество пользователей в каждой когорте
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head()
```

Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждую следующую неделю.

```python
# Найдём исходное количество пользователей в когорте на нулевую неделю
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]

# Переименуем столбец uid в cohort_users
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

# Объединим данные по когортам с исходным количеством пользователей в когорте
cohorts = cohorts.merge(initial_users_count,on='first_visit_month')

# считаем Retention Rate
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
cohorts
```

Построим сводную таблицу и создадим тепловую карту.

```python
retention_pivot = cohorts.pivot_table(
                            index='first_visit_month',
                            columns='cohort_lifetime',
                            values='retention',
                            aggfunc='sum')
retention_pivot.fillna('')
```

Так как видим, что retention во всех месяцах, начиная с первого, очень низкий, то построим тепловую карту, **отбросив нулевой месяц lifetime**, иначе контраст между нулевым и всеми остальными месяцами будет такой большой, что будет плохо видно тренды.

```python
# убираем нулевой лайфтайм (только для визуализации)
retention_pivot_2 = retention_pivot.drop(0, axis=1)
# добавляем расчет среднего по времени жизни среди всех когорт
retention_pivot_2.loc['mean'] = retention_pivot_2.mean()

# строим heatmap с помощью библиотеки seaborn
sns.set(style='white')
plt.figure(figsize=(15, 9))
plt.title('Retention Rate по когортам')
sns.heatmap(retention_pivot_2, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
```

Найдем средний Retention Rate на второй месяц «жизни» когорт.

```python
cohorts[cohorts['cohort_lifetime'] == 1]['retention'].mean()
```

Всего около 6,5% пользователей возвращаются на второй месяц! 

**Вывод**

- Retention очень сильно упал в апреле и мае по всем когортам, независимо от lifetime - что-то пошло не так?
- Наихудший retention наблюдаем в когорте марта 2018 - в первый месяц вернулось всего 2.5% пользователей
- Практически во всех когортах показатель колеблется то в худшую, то в лучшую сторону, нет четкого тренда на понижение, кроме последних двух месяцев. Хорошо бы разобраться, что случилось в этот период.
- Средний Retention на второй месяц жизни когорт 6.5%, то есть в среднем возвращаются на второй месяц всего лишь 6.5% пользователей..

Мы ранее уже определили для каждого юзера дату первого визита: она находится в таблице `visits` в атрибуте `first_visit`. Выделим только необходимые для расчета столбцы в новый датафрейм `first_visits2`

```python
first_visits2 = visits[['uid', 'first_visit']]
first_visits2.head()
```

Теперь в таблице `orders` посчитаем дату первой покупки.

```python
first_order = orders.groupby('uid')['buy_ts'].min().reset_index()
first_order.head()
```

Соединяем эти два датафрейма в один и считаем разность между датой первой покупки и датой первого визита.

```python
# мержим
first_orders = first_visits2.merge(first_order, on='uid', how='right')
first_orders

# сделаем названия столбцов более понятными.
first_orders = first_orders.rename(columns={'first_visit': 'first_visit_date', 'buy_ts': 'first_order_date'})
```

```python
# оставляем только когорты, прожившие дольше 6 месяцев
# и заказы, сделанные за время жизни <=6, т.е. delta <= 180 ?
first_orders2 = first_orders.query('first_visit_month < "2017-12-01" & first_purchase_delta <= 180')
```

#### 2.2.2 Cреднее количество покупок на одного покупателя

- Сколько раз покупают за определённый период, например за 6 месяцев? 

Чтобы ответить на этот вопрос, снова применим когортный анализ. На этот раз когорты сформируем по дате первой покупки.

```python
# выделяем из даты покупки месяц
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

# группируем и ищем дату первой покупки для каждого покупателя
first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

#Построим когорты. Добавим месяц первой покупки каждого покупателя в таблицу с заказами:
orders_new = pd.merge(orders, first_orders, on='uid')
orders_new.head()
```

Формируем когорты.

```python
#Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и посчитаем количество заказов всего
# и количество уникальных пользователей, совершивших эти заказы. 
cohorts = orders_new.groupby(['first_order_month', 'order_month']).agg({'uid': 'count'}).reset_index()
cohorts.columns=['first_order_month', 'order_month', 'n_orders']
cohorts.head()
```

Считаем количество покупателей в каждой когорте (размер когорты)

```python
# группировка по месяцу когорты и считаем количество уникальных uid
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes.head()
```

Добавляем размер когорт в таблицу `cohorts`

```python
cohorts = cohorts.merge(cohort_sizes, on='first_order_month')
cohorts
```

Теперь, чтобы рассчитать среднее количество покупок за 6 месяцев, добавим в `cohorts` время "жизни" когорты, т.е. `lifetime`.

```python
cohorts['lifetime'] = (cohorts['order_month'] - cohorts['first_order_month']) / np.timedelta64(1, 'M')
cohorts['lifetime'] = cohorts['lifetime'].round().astype('int')
cohorts.head()
```

Посчитаем количество покупок на 1 пользователя в каждой когорте и запишем в столбец `orders_per_user`

```python
cohorts['orders_per_user'] = cohorts['n_orders'] / cohorts['n_buyers']

#округляем до 2 знаков после запятой
cohorts['orders_per_user'] = cohorts['orders_per_user'].round(2)
```

Построим сводную таблицу по когортам и количеству заказов.

```python
orders_per_user_pivot = cohorts.pivot_table(
                                index='first_order_month',
                                columns='lifetime',
                                values='orders_per_user',
                                aggfunc='mean')
orders_per_user_pivot.fillna('')
```

Для того чтобы узнать среднее количество покупок на пользователя за 6 месяцев, посчитаем накопительную сумму по когортам и выделим столбец с 5-м месяцем lifetime.

```python
res = orders_per_user_pivot.cumsum(axis=1)

sns.set(style='white')
plt.figure(figsize=(15, 9))
plt.title('Среднее количество покупок на покупателя по когортам')
res.index = res.index.strftime('%Y-%m')
sns.heatmap(res, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
```

Если смотреть на столбец lifetime 5, то самая высокая накопительная сумма приходится на первую когорту: 1.69 заказов за 6 месяцев на одного покупателя. Меньше всего заказов за 6 месяцев в октябрьской и декабрьской когортах.

Посчитаем среднее количество заказов на 6 месяцев по всем когортам.

```python
avg_orders = round(orders_per_user_pivot.cumsum(axis=1)[5].mean(), 2)
print('Среднее количество заказов на одного покупателя за 6 месяцев: ', avg_orders)
```

**Вывод**
- Наибольшее количество покупок на одного покупателя приходится на самую первую когорту (июнь 2017): 1.69. 
- Самый низкий показатель у октябрьской когорты: за 6 месяцев 1.33 покупок на пользователя.
- В среднем за 6 месяцев один покупатель совершил 1.43 покупок.

- В первый месяц "жизни" когорты средний чек обычно чуть ниже 5 у.е.


#### 2.2.4 Накопительный LTV по когортам

- Выясним, как меняется во времени LTV на покупателя по когортам.

Маржинальность сервиса — 100%, поэтому будем брать за прибыль просто столбец `revenue`. Все необходимые данные для расчета `ltv` уже есть в таблице `report`. Формула такая: делим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте.

```python
report['ltv'] = report['revenue'] / report['n_buyers']

report.head()
```

```python
#Для наглядности визуализируем когорты в виде сводной таблицы:
output = report.pivot_table(
            index='first_order_month',
            columns='lifetime',
            values='ltv',
            aggfunc='sum').round(2)

output.index = output.index.strftime('%Y-%m')

# среднее значение за время жизни среди всех когорт
output.loc['mean'] = output.mean()

# LTV - накопительный показатель, поэтому выведем накопительные суммы
output.cumsum(axis=1).fillna('')
```

В строках таблицы мы видим накопительный LTV по каждой когорте. Например, для когорты сентября накопительный LTV на конец расчетного периода составил 13.44 у.е. с 1 покупателя.

- Рассчитаем средний накопительный LTV по всем когортам вместе взятым за 6 месяцев. В расчёт войдут когорты, «прожившие» не менее 6 месяцев, т.е когорты, сформированные не позже декабря, т.к. по остальным когортам в этом столбце будут пропуски.

```python
output.cumsum(axis=1)[5].mean()
```

В среднем за 6 месяцев когорты принесли по 7.977 у.е. прибыли.

**Вывод**
- Мы посчитали накопительный LTV по когортам и усредненный LTV за 6 месяцев.
- Самый высокий LTV за 6 месяцев принесла сентябрькая когорта: 12.32 у.е.
- Как правило, большинство покупок совершается в первый месяц жизни когорт, поэтому накопительный LTV прирастает медленно, однако все же растет, что очень хорошо: это значит, пользователи возвращаются и совершают повторные покупки.

#### 2.3.3 ROMI по когортам в разрезе источников 


Данные о дате первой покупки и выручке у нас уже сагрегированы в таблице `avg_check`

```python
avg_check.head()
```

Теперь в эту таблицу нужно добавить главный источник для каждого `uid`. Эти данные уже есть в `users`

```python
# сразу отфильтровываем только нужные нам столбцы
user_source = users[['uid', 'source_id','first_visit']]

orders_new = avg_check.merge(user_source, on='uid')
```

Теперь для формирования когорт нам нужен месяц первого заказа.

```python
# считаем дату и время первого заказа
first_order_dt = orders.groupby('uid')['buy_ts'].min().reset_index()
first_order_dt.columns = ['uid', 'first_order_dt']

# мержим с основной таблицей
orders_new = orders_new.merge(first_order_dt, on='uid')
orders_new.head()
```

Так как LTV и ROMI накопительный будем сравнивать по одинаковому времени "жизни" когорт, то еще пригодится lifetime

```python
# считаем время "жизни"
orders_new['lifetime'] = (orders_new['order_month'] - orders_new['first_order_month']) 
orders_new['lifetime'] = orders_new['lifetime'] / np.timedelta64(1, 'M')
orders_new['lifetime'] = orders_new['lifetime'].round().astype('int')
#orders_new.head(50)
```

Формируем когорты и считаем валовую (накопительную) прибыль по когортам.

```python
revenues = orders_new.groupby(['first_order_month', 'lifetime', 'source_id'])['revenue'].sum().groupby(['first_order_month', 'source_id']).cumsum().reset_index()

# check
revenues.query('first_order_month == "2017-12-01" & source_id == 1')
```

Для расчета **ltv** также понадобится количество пользователей в каждой когорте и источнике. Эти данные у нас уже есть в таблице `sources` выше.

```python
cac = sources[['source_id', 'month', 'users', 'cac']]
cac.columns = ['source_id', 'first_order_month', 'cohort_size', 'cac']

# склеиваем доходы и расходы в одну таблцу
ltv = revenues.merge(cac, on=['first_order_month', 'source_id'])
```

Теперь считаем ltv

```python
# считаем по формуле, выручка на количество покупателей в когорте
ltv['ltv'] = ltv['revenue'] / ltv['cohort_size']
```

Для расчета ROMI нам нужны будут также значения САС. Они у нас уже расчитаны в предыдущем пункте. Далее считае возврат на маркетинговые инвестиции по формуле `ROMI = LTV / CAC`

```python
# считаем ROMI
ltv['romi'] = ltv['ltv'] / ltv['cac']
ltv
```

Уже можно посмотреть средние LTV и CAC по источникам

```python
ltv.groupby('source_id')['ltv', 'cac'].mean()
```

По среднему итогу, не все когорты окупились. Для 2, 3, 4, 10 источников cac > ltv, т.е прибыль не покрыла расходов на маркетинг.
